Compare the whole dataset vs na dataset by showing the top k popular item graph

```python
import codecs
import csv
from collections import Counter
import pandas as pd
```

```python
#use na dataset
```

```python
list_=[]
with codecs.open('merge_table_drop_na.csv', encoding='utf-8-sig') as f:
    for row in csv.DictReader(f, skipinitialspace=True):
        list_.append(row["item_id"])
```

```python
cnt = Counter(list_).most_common()
```

```python
cnt[0:30]
```

```python
import pandas as pd
import matplotlib.pyplot as plt	
```

```python
def draw(s,k):
    x_label=[s[i][0] for i in range(k)]
    y_label=[s[i][1] for i in range(k)]
    df = pd.DataFrame({'item_id':x_label, 'frequency':y_label})
    ax = df.plot.bar(x='item_id', y='frequency', rot=1000)
```

```python
draw(cnt,10),draw(cnt,20),draw(cnt,30) #top 10,20,30 popular
```

```python
#use whole dataset
```

```python
list_2=[]
with codecs.open('user_item_review_merge_table.csv', encoding='utf-8-sig') as f:
    for row in csv.DictReader(f, skipinitialspace=True):
        list_2.append(row["item_id"])
```

```python
cnt2 = Counter(list_2).most_common()
```

```python
cnt2[0:30]
```

```python
draw(cnt2,10),draw(cnt2,20),draw(cnt2,30) #top 10,20,30 popular
```
