# Data Preprocessing

```python
import gzip
import csv
f = open("australian_users_items.json",encoding='utf-8')
dataset = []
for l in f:
    dataset.append(eval(l))
```

# user_item.csv

```python
import pandas as pd
```

```python
df_dict = {'user_id': [], 'item_id': [], 'item_name': [], 'playtime_forever': [] }

for user_index in range(len(dataset)):
    user_id = dataset[user_index]['user_id']
    items_length = len(dataset[user_index]['items']) 
    
    for item_index in range(items_length):
        df_dict['user_id'].append(user_id)
        item_id = dataset[user_index]['items'][item_index]['item_id']
        item_name = dataset[user_index]['items'][item_index]['item_name']
        playtime = dataset[user_index]['items'][item_index]['playtime_forever']
        # append
        df_dict['item_id'].append(item_id)
        df_dict['item_name'].append(item_name)
        df_dict['playtime_forever'].append(playtime)
        
    
```

```python
df = pd.DataFrame(data=df_dict)
```

# user_review.csv

```python
import gzip
import csv
f = open("australian_user_reviews.json",encoding='utf-8')
dataset = []
for l in f:
    dataset.append(eval(l))
```

```python
dataset[0]['reviews'][0]
```

```python
import pandas as pd

df_dict = {'user_id': [], 'item_id': [], 'recommend': [], 'review': []}

for user_index in range(len(dataset)):
    user_id = dataset[user_index]['user_id']
    reviews_length = len(dataset[user_index]['reviews']) 
    
    for review_index in range(reviews_length):
        df_dict['user_id'].append(user_id)
        item_id = dataset[user_index]['reviews'][review_index]['item_id']
        recommend = dataset[user_index]['reviews'][review_index]['recommend']
        review = dataset[user_index]['reviews'][review_index]['review']
        # append
        df_dict['item_id'].append(item_id)
        df_dict['recommend'].append(recommend)
        df_dict['review'].append(review)
```

```python
df = pd.DataFrame(data=df_dict)
```

```python
df.to_csv('user_review_.csv',index=False)
```
